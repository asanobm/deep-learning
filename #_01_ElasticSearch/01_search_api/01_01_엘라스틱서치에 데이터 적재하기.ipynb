## 1.1. 엘라스틱서치에 데이터 적재하기

```python
import json
import pandas as pd
from elasticsearch import Elasticsearch
client = Elasticsearch(
    hosts=['https://localhost:9200'],
    http_auth=('elastic', 'qwer1234!'),
    verify_certs=False
)

print(json.dumps(client.info().body, indent=2))
```

```python
indexes = client.cat.indices(format='json')
df = pd.DataFrame(indexes)
df
```

```python
naver_place_docs = client.search(index='naver_place', body={"query": {"match_all": {}}})
naver_place_df = pd.json_normalize(naver_place_docs['hits']['hits'])
naver_place_df.shape
```
