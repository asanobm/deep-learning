### 3.4.2. íŠ¸ë ˆì´ë„ˆ APIë¥¼ ì‚¬ìš©í•´ í•™ìŠµí•˜ê¸°

Hugging Faceì˜ Trainer APIë¥¼ ì‚¬ìš©í•´ ëª¨ë¸ì„ í•™ìŠµí•œë‹¤.

````python
import torch
import torch.nn as nn

from datasets import load_dataset
from transformers import Trainer, TrainingArguments, AutoModelForSequenceClassification, AutoTokenizer

# KLUE-YNAT ë°ì´í„°ì…‹ì„ ë¶ˆëŸ¬ì˜¨ë‹¤. 
dataset = load_dataset("klue", "ynat")


device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')

# Split the dataset into train, validation, and test sets
train_test_split = dataset["train"].train_test_split(test_size=0.1, shuffle=True, seed=42)
train_dataset = train_test_split["train"]
eval_dataset = train_test_split["test"]
test_dataset = dataset["validation"].train_test_split(test_size=1000, shuffle=True, seed=42)["test"]

# ì‚¬ìš©í•˜ì§€ ì•ŠëŠ” ì»¬ëŸ¼ ì œê±°
train_dataset = train_dataset.remove_columns(['guid', 'url', 'date'])
eval_dataset = eval_dataset.remove_columns(['guid', 'url', 'date'])
test_dataset = test_dataset.remove_columns(['guid', 'url', 'date'])


# í† í¬ë‚˜ì´ì €ì™€ ëª¨ë¸ ì„¤ì •
tokenizer = AutoTokenizer.from_pretrained("klue/bert-base")
model = AutoModelForSequenceClassification.from_pretrained(
    "klue/bert-base", # KLUE-BERT ëª¨ë¸ ì‚¬ìš©
    num_labels=dataset['train'].features['label'].num_classes # í´ë˜ìŠ¤ ê°œìˆ˜ë¥¼ ì„¤ì •í•œë‹¤. ì‹¤ì œ ë°ì´í„°ì…‹ì˜ í´ë˜ìŠ¤ ê°œìˆ˜ì™€ ê°™ì•„ì•¼ í•œë‹¤.
)

# tokenize_function ì€ ë°ì´í„°ì…‹ì˜ exampleì„ í† í°ìœ¼ë¡œ ë³€í™˜í•œë‹¤. ë³€í™˜í•˜ê³ ìí•˜ëŠ” í•„ë“œëŠ” titleì´ë‹¤. 
def tokenize_function(examples):
    """
    ì£¼ì–´ì§„ ì˜ˆì œì˜ 'title' í•„ë“œë¥¼ í† í¬ë‚˜ì´ì €ë¥¼ ì‚¬ìš©í•˜ì—¬ í† í°í™”í•©ë‹ˆë‹¤.
    Args:
        examples (dict): 'title' í•„ë“œë¥¼ í¬í•¨í•˜ëŠ” ì˜ˆì œë“¤ì˜ ë”•ì…”ë„ˆë¦¬.
    Returns:
        dict: íŒ¨ë”©ê³¼ ì˜ë¦¼ì´ ì ìš©ëœ í† í°í™”ëœ ê²°ê³¼ë¥¼ í¬í•¨í•˜ëŠ” ë”•ì…”ë„ˆë¦¬.
    """
    return tokenizer(examples['title'], padding="max_length", truncation=True)

# tokenize_functionì„ ì‚¬ìš©í•˜ì—¬ ë°ì´í„°ì…‹ì„ í† í°í™”í•œë‹¤.
train_dataset = train_dataset.map(tokenize_function, batched=True)
eval_dataset = eval_dataset.map(tokenize_function, batched=True)
test_dataset = test_dataset.map(tokenize_function, batched=True)

# TrainingArgumentsë¥¼ ì‚¬ìš©í•˜ì—¬ í•™ìŠµ ì„¤ì •ì„ ì •ì˜í•œë‹¤.
training_args = TrainingArguments(
    output_dir='./runs',                # ëª¨ë¸ ì²´í¬í¬ì¸íŠ¸ ë° ì˜ˆì¸¡ ê²°ê³¼ë¥¼ ì €ì¥í•  ë””ë ‰í† ë¦¬
    num_train_epochs=1,                 # í•™ìŠµí•  ì—í­ ìˆ˜
    per_device_train_batch_size=8,      # í•œ ë””ë°”ì´ìŠ¤ë‹¹ í•™ìŠµ ë°°ì¹˜ í¬ê¸°
    per_device_eval_batch_size=8,       # í•œ ë””ë°”ì´ìŠ¤ë‹¹ í‰ê°€ ë°°ì¹˜ í¬ê¸°
    evaluation_strategy="epoch",        # í‰ê°€ ì „ëµ
    learning_rate=5e-5,                 # í•™ìŠµë¥ 
)

def compute_metrics(pred):
    """
    ì£¼ì–´ì§„ ì˜ˆì¸¡ ê²°ê³¼ë¥¼ ë°”íƒ•ìœ¼ë¡œ ì •í™•ë„ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.
    ë§¤ê°œë³€ìˆ˜:
    pred (transformers.EvalPrediction): ì˜ˆì¸¡ ê²°ê³¼ë¥¼ í¬í•¨í•˜ëŠ” ê°ì²´ë¡œ, 
        label_idsì™€ predictions ì†ì„±ì„ ê°€ì§‘ë‹ˆë‹¤.
    ë°˜í™˜ê°’:
    dict: ì •í™•ë„ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë”•ì…”ë„ˆë¦¬ë¡œ, "accuracy" í‚¤ì™€ ì •í™•ë„ ê°’ì„ ê°€ì§‘ë‹ˆë‹¤.
    """

    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    return {"accuracy": (preds == labels).mean()}

# Trainer ê°ì²´ë¥¼ ìƒì„±í•œë‹¤.
trainer = Trainer(
    model=model,                    # í•™ìŠµí•  ëª¨ë¸
    args=training_args,             # í•™ìŠµ ì„¤ì •
    train_dataset=train_dataset,    # í•™ìŠµ ë°ì´í„°ì…‹
    eval_dataset=eval_dataset,      # í‰ê°€ ë°ì´í„°ì…‹
    tokenizer=tokenizer,            # í† í¬ë‚˜ì´ì €
    compute_metrics=compute_metrics # í‰ê°€ ì§€í‘œë¥¼ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜
)

# ëª¨ë¸ì„ DataParallelë¡œ ê°ì‹¸ê¸° : ì—¬ëŸ¬ GPUë¥¼ ì‚¬ìš©í•˜ì—¬ í•™ìŠµí•  ë•Œ ì‚¬ìš© í•´ì•¼ í•œë‹¤.
if torch.cuda.device_count() > 1:
    print("Let's use", torch.cuda.device_count(), "GPUs!")
    model = nn.DataParallel(model)

model.to(device)

# ëª¨ë¸ì„ í•™ìŠµí•œë‹¤.
trainer.train()
````

````python
import json
metrics = trainer.evaluate(test_dataset)
print(json.dumps(metrics, indent=2))
````

### 3.4.3. íŠ¸ë ˆì´ë„ˆ APIë¥¼ ì‚¬ìš©í•˜ì§€ ì•Šê³  í•™ìŠµí•˜ê¸°

Trainerë¥¼ ì‚¬ìš©í•˜ë©´ ê°„í¸í•˜ê²Œ ì‚¬ìš©í•  ìˆ˜ ìˆì§€ë§Œ ë‚´ë¶€ ë™ì‘ì„ ì´í•´í•˜ê¸° ìœ„í•´ Trainerë¥¼ ì‚¬ìš©í•˜ì§€ ì•Šê³  í•™ìŠµí•˜ëŠ” ë°©ë²•ì„ ì•Œì•„ë³¸ë‹¤.

````python
import torch
import torch.nn as nn
import numpy as np
from tqdm.auto import tqdm
from torch.utils.data import DataLoader
from transformers import AdamW
from datasets import load_dataset

def tokenize_function(examples):
    return tokenizer(examples['title'], padding="max_length", truncation=True)

# KLUE-YNAT ë°ì´í„°ì…‹ì„ ë¶ˆëŸ¬ì˜¨ë‹¤.
dataset = load_dataset("klue", "ynat")

# tokenizerì™€ modelì„ ì„¤ì •í•œë‹¤.
tokenizer = AutoTokenizer.from_pretrained("klue/bert-base")
model = AutoModelForSequenceClassification.from_pretrained("klue/bert-base",num_labels=dataset['train'].features['label'].num_classes)

# ì‚¬ìš©í•  ë””ë°”ì´ìŠ¤ë¥¼ ì„¤ì •í•œë‹¤.
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')

# ëª¨ë¸ì„ DataParallelë¡œ ê°ì‹¸ê¸° : ì—¬ëŸ¬ GPUë¥¼ ì‚¬ìš©í•˜ì—¬ í•™ìŠµí•  ë•Œ ì‚¬ìš© í•´ì•¼ í•œë‹¤.
if torch.cuda.device_count() > 1:
    print("Let's use", torch.cuda.device_count(), "GPUs!")
    model = nn.DataParallel(model)
model.to(device)

def make_data_loader(dataset, batch_size, shuffle=True):
    dataset = dataset.map(tokenize_function, batched=True).with_format("torch")
    dataset = dataset.rename_column("label", "labels")
    dataset = dataset.remove_columns(['title'])
    return DataLoader(dataset, batch_size=batch_size, shuffle=shuffle)

# ë°ì´í„°ë¡œë”ë¥¼ ìƒì„±í•œë‹¤.
train_loader = make_data_loader(train_dataset, batch_size=8, shuffle=True)
eval_loader = make_data_loader(eval_dataset, batch_size=8, shuffle=False)
test_loader = make_data_loader(test_dataset, batch_size=8, shuffle=False)


# í•™ìŠµì„ ìœ„í•œ í•¨ìˆ˜ë¥¼ ì •ì˜í•œë‹¤.
def train_epoch(model, data_loader, optimizer):
    # ëª¨ë¸ì„ í•™ìŠµ ëª¨ë“œë¡œ ì„¤ì •í•œë‹¤.
    model.train()
    # ì „ì²´ ì†ì‹¤ì„ ì €ì¥í•  ë³€ìˆ˜ë¥¼ ì´ˆê¸°í™”í•œë‹¤.
    total_loss = 0
    
    # ë°ì´í„°ë¡œë”ì—ì„œ ë°°ì¹˜ë¥¼ ìˆœíšŒí•˜ë©° í•™ìŠµí•œë‹¤.
    for batch in tqdm(data_loader):
        # optimizerì˜ gradientë¥¼ ì´ˆê¸°í™”í•œë‹¤.
        optimizer.zero_grad()
        # input_idsë¥¼ GPUë¡œ ì˜®ê¸´ë‹¤.
        input_ids = batch['input_ids'].to(device)
        # attention_maskë¥¼ GPUë¡œ ì˜®ê¸´ë‹¤.
        attention_mask = batch['attention_mask'].to(device)
        # labelsë¥¼ GPUë¡œ ì˜®ê¸´ë‹¤.
        labels = batch['labels'].to(device)
        # ëª¨ë¸ì— input_ids, attention_mask, labelsë¥¼ ì „ë‹¬í•˜ì—¬ ì¶œë ¥ì„ ê³„ì‚°í•œë‹¤.
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        # ì†ì‹¤ì„ ê³„ì‚°í•œë‹¤.
        loss = outputs.loss
        # ë§Œì•½ ì†ì‹¤ì´ scalarê°€ ì•„ë‹ˆë¼ë©´, í‰ê· ì„ ê³„ì‚°í•œë‹¤.
        if loss.dim() > 0:
            loss = loss.mean()
        # ì†ì‹¤ì„ ì—­ì „íŒŒí•˜ì—¬ gradientë¥¼ ê³„ì‚°í•œë‹¤.
        loss.backward()
        # optimizerë¥¼ ì‚¬ìš©í•˜ì—¬ ëª¨ë¸ íŒŒë¼ë¯¸í„°ë¥¼ ì—…ë°ì´íŠ¸í•œë‹¤.
        optimizer.step()
        # ì†ì‹¤ì„ ê¸°ë¡í•œë‹¤.
        total_loss += loss.item()
    avg_loss = total_loss / len(data_loader)
    return avg_loss

# í‰ê°€ë¥¼ ìœ„í•œ í•¨ìˆ˜ë¥¼ ì •ì˜í•œë‹¤.
def evaluate(model, data_loader):
    # ëª¨ë¸ì„ í‰ê°€ ëª¨ë“œë¡œ ì„¤ì •í•œë‹¤.
    model.eval()
    # ì „ì²´ ì†ì‹¤ì„ ì €ì¥í•  ë³€ìˆ˜ë¥¼ ì´ˆê¸°í™”í•œë‹¤.
    total_loss = 0
    # ì •í™•ë„ë¥¼ ì €ì¥í•  ë³€ìˆ˜ë¥¼ ì´ˆê¸°í™”í•œë‹¤.
    predictions = []
    # ì •ë‹µì„ ì €ì¥í•  ë³€ìˆ˜ë¥¼ ì´ˆê¸°í™”í•œë‹¤.
    true_labels = []

    # gradientë¥¼ ê³„ì‚°í•˜ì§€ ì•Šë„ë¡ ì„¤ì •í•œë‹¤.
    with torch.no_grad():
        for batch in tqdm(data_loader):
            # input_idsë¥¼ GPUë¡œ ì˜®ê¸´ë‹¤.
            input_ids = batch['input_ids'].to(device)
            # attention_maskë¥¼ GPUë¡œ ì˜®ê¸´ë‹¤.
            attention_mask = batch['attention_mask'].to(device)
            # labelsë¥¼ GPUë¡œ ì˜®ê¸´ë‹¤.
            labels = batch['labels'].to(device)
            # ëª¨ë¸ì— input_ids, attention_maskë¥¼ ì „ë‹¬í•˜ì—¬ ì¶œë ¥ì„ ê³„ì‚°í•œë‹¤.
            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
            # ë¡œì§“ì„ ê³„ì‚°í•œë‹¤.
            logits = outputs.logits
            # ì†ì‹¤ì„ ê³„ì‚°í•œë‹¤.
            loss = outputs.loss
            # ë§Œì•½ ì†ì‹¤ì´ scalarê°€ ì•„ë‹ˆë¼ë©´, í‰ê· ì„ ê³„ì‚°í•œë‹¤.
            if loss.dim() > 0:
                loss = loss.mean()
            # ì†ì‹¤ì„ ê¸°ë¡í•œë‹¤.
            total_loss += loss.item()
            # í…ì„œ logitsì˜ ê° í–‰ì—ì„œ ìµœëŒ“ê°’ì˜ ì¸ë±ìŠ¤ë¥¼ ì°¾ê³ , ê·¸ ê²°ê³¼ë¥¼ predsì— ì €ì¥.
            preds = torch.argmax(logits, dim=-1)
            # predictionsì— preds ê²°ê³¼ë¥¼ cpuë¡œ ì˜®ê²¨ì„œ ì €ì¥.
            predictions.extend(preds.cpu().numpy())
            # true_labelsì— labels ê²°ê³¼ë¥¼ cpuë¡œ ì˜®ê²¨ì„œ ì €ì¥.
            true_labels.extend(labels.cpu().numpy())
    
    avg_loss = total_loss / len(data_loader)
    accuracy = np.mean(predictions == true_labels) == np.array(true_labels)
    return avg_loss, accuracy

num_epochs = 1
optimizer = AdamW(model.parameters(), lr=5e-5)

for epoch in range(num_epochs):
    print(f"Epoch {epoch +1}/{num_epochs}")
    train_loss = train_epoch(model, train_loader, optimizer)
    print(f"Train loss: {train_loss}")
    valid_loss, valid_acc = evaluate(model, eval_loader)
    print(f"Validation loss: {valid_loss}, Validation accuracy: {valid_acc}")
    _, test_acc = evaluate(model, test_loader)
    print(f"Test accuracy: {test_acc}")
````
