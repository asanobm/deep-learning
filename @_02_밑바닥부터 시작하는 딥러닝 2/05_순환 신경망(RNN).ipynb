# 5 순환 신경망(RNN)

지금까지 살펴본 신경망(? 행렬곱? 소프트맥스? 시그모이드? word2vec?, CBOW?)은 피드 포워드(feed forward)라는 유형이다. 피드포워드란 흐름이 단방향인 신경망을 말한다. 즉, 입력층에서 출력층으로의 연산이 한 방향으로만 흐른다.

피드포워드는 단순하고 구조가 이해하기 쉽지만 `시계열 데이터`를 잘 다루지 못한다. (안타깝다)

시계열 데이터란 시간 순서대로 등장하는 데이터를 말한다. 예를 들어, 주식 가격, 날씨, 자연어 등이 시계열 데이터에 속한다. 이러한 시계열 데이터를 잘 다루기 위해 순환 신경망(Recurrent Neural Network, RNN)이라는 신경망이 등장했다.

## 5.1 확률과 언어 모델

### 5.1.1 word2vec의 문제점

$w_{1}, w_{2}, ... w_{T}$라는 단어열로 표현된 말뭉치가 있다고 가정하고 t번째 단어를 타깃으로 했을 때, word2vec은 이전 단어들을 맥락으로 이용해 t번째 단어의 출현 확률을 구한다.
즉, t를 추측하기 위해 $w_{t-1}, w_{t+1}$을 이용한다.

이를 수식으로 나타내면 다음과 같다.
$$
P(w_{t}|w_{t-1}, w_{t-2}, ... w_{t-n+1})
$$

이것이 윈도우 크기가 1일 때 CBOW모델이다.

만약 왼쪽의 단어만을 이용해 t번째 단어를 추측하고 싶다면$P(w_{t}|w_{t-2}, w_{t-1})$, 
손실 함수는 $L = -\log P(w_{t}|w_{t-2}, w_{t-1})$이다.
