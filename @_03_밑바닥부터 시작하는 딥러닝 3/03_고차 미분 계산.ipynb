## 26 계산 그래프 시각화 (2)


### 26.1 시각화 코드

```python
import numpy as np
from dezero import Variable
from dezero.utils import plot_dot_graph

x0 = Variable(np.array(1.0))
x1 = Variable(np.array(1.0))
y = x0 + x1

x0.name = 'x0'
x1.name = 'x1'
y.name = 'y'

txt = get_dot_graph(y, verbose=False)
print(txt)
```

이런식으로 get_dot_graph 을 사용하면 계산 그래프를 시각화 할 수 있습니다.
y는 Variable 클래스의 인스턴스이므로 계산 그래프를 생성할 수 있습니다.

````python
import numpy as np
import graphviz
from framework import Variable, plot_dot_graph, get_dot_graph

def goldstein(x, y):
  z = (1 + (x + y + 1)**2 * (19 - 14*x + 3*x**2 - 14*y + 6*x*y + 3*y**2)) * \
      (30 + (2*x - 3*y)**2 * (18 - 32*x + 12*x**2 + 48*y - 36*x*y + 27*y**2))
  return z

x = Variable(np.array(1.0))
y = Variable(np.array(1.0))
z = goldstein(x, y)
z.name = 'z'

x.name = 'x'
y.name = 'y'
z.name = 'z'

# plot_dot_graph(z, verbose=False, to_file='goldstein.png')

source = get_dot_graph(z, verbose=False)
graphviz.Source(source)
````
