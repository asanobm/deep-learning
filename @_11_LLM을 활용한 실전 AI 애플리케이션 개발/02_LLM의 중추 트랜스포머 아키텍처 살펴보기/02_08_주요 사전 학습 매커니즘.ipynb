## 2.8. 주요 사전학습 메커니즘

### 2.8.1. 인과적 언어 모델링

인과적 언어 모델링은 문장의 시작부터 끝까지 순차적으로 단어를 예측하는 방식이다. 이전에 등장한 단어들을 바탕을 다음에 등장할 단어를 예측한다. 이러한 방식은 사람이 문장을 만드는 방식과 유사하다. 예를 들어 '나는 밥을 먹었다'라는 문장을 만들 때 '나는'이라는 단어를 먼저 생각하고 그 다음에 '밥을'이라는 단어를 생각하고 그 다음에 '먹었다'라는 단어를 생각한다. 이러한 방식으로 문장을 만들어 나가는 것이 인과적 언어 모델링이다. 

### 2.8.2. 마스크 언어 모델링

마스크 언어 모델링은 씨끄러운 공간에서 대화를 나누는 경우를 상상해보자. 시끄러운 상황에서 대화를 나누면 중간중간에 몇개의 단어는 들리지 않는다. 익숙한 단어가 몇개 들리지 않아도 문맥을 이해하고 대화를 이어나갈 수 있다. 마스크 언어 모델링은 이러한 상황을 모델링한 것이다. 문장의 일부를 가리고 나머지 문장을 보고 가려진 부분을 예측하는 방식이다. 이러한 방식은 문장의 일부가 가려져 있어도 문맥을 이해하고 문장을 이어나갈 수 있다.
