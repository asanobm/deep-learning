# 컴퓨터는 데이터에서 배운다.

* 머신 러닝의 일반적 개념
* 세 종류의 학습과 기본 용어
* 성공적인 머신러닝 시스템을 설계하는 필수 요소
* 화이팅

## 1.1. 데이터를 지식으로 바꾸는 지능적인 시스템 구축

* 머신 러닝은 데이터에서 지식을 추출하는 방법을 연구하는 분야
* 머신 러닝은 데이터를 이용해 모델을 만들고, 이 모델을 사용해 새로운 데이터에 대한 예측을 수행
* 돈을 벌거나, 과학적인 발견을 하는 등 다양한 목적으로 사용

## 1.2. 머신 러닝의 3가지 종류

* 지도 학습: 출력 및 미래 예측
* 비지도 학습: 숨겨진 구조 및 패턴 발견
* 강화 학습: 연속된 행동에서 학습

### 1.2.1. 지도 학습
지도 학습의 주요 목적은 레이블된 훈련 데이터에서 모델을 학습하여 본적 없는 미래 데이터에 대해 예측하는 것이다.

**분류:클래스 레이블 예측**
과거의 관측을 기반으로 새로운 샘플의 범주형 클래스 레이블을 예측하는 것이 목적이다.
스팸 메일 필터링은 이진 분류 작업의 예이다.

**결정 경계:클래스를 나누는 경계**
분류 작업을 수행할 때 결정 경계를 찾아야 한다. 결정 경계는 클래스를 나누는 경계이다.

**다중 분류:3개 이상의 클래스 레이블 예측**
3개 이상의 클래스 레이블을 예측하는 것이 목적이다. 예를 들어, 손글씨 숫자 분류는 다중 분류 작업의 예이다.

**회귀:연속적인 출력 값 예측**
연속적인 출력 값을 예측하는 것이 목적이다. 회귀는 *예측 변수(설명 변수)*와 연속적인 *반응 변수*가 주어졌을 때 출력 값을 예측하기 위해 두 변수 사이의 관계를 찾는다.

**선형 회귀:선형 모델을 사용한 회귀**
특성x와 타깃y가 주어지면 데이터 포인트와 직선 사이 거리가 최소가 되는 직선을 그을 수 있다.
일반적으로 평균 제곱 거리를 사용한다.
데이터에서 학습한 절편을 사용하여 새로운 데이터의 출려값을 예측한다.

### 1.2.2. 강화학습으로 반응형 문제 해결

강화학습은 시스템이 환경에 대한 정보를 받아 보상을 최대화하는 행동을 선택하는 방법을 학습하는 것이다.
강화학습의 피드백은 정답이레이블이나 값이 아니라 보상 함수로 얼마나 행동이 좋은지를 측정한 값이다.
에이전트는 환경과 상호 작용하여 보상이 최대화되는 일련의 행동을 학습한다.

### 1.2.3. 비지도 학습으로 숨겨진 구조 발견

비지도 학습은 레이블이 없는 데이터에서 구조를 발견하는 것이다.

**군집:서브그룹 찾기**

*군집*은 그룹 소속에 대한 사전 정보 없이 쌓여 있는 데이터를 의미 있는 *서브그룹* 또는 *클러스터*로 조직하는 탐색적 데이터 분석 또는 패턴 발견 기법이다.

**차원 축소:데이터 압축**

*차원 축소*는 데이터의 특성 수를 줄이는 것이다. 차원 축소는 데이터를 시각화하거나 노이즈를 제거하거나 데이터를 압축하는 데 사용된다.
